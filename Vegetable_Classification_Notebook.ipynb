```python
# Directories for train, validation, test data

train_data_dir = r'D:\data_projects\veggie_classification\train'
val_data_dir = r'D:\data_projects\veggie_classification\validation'
test_data_dir = r'D:\data_projects\veggie_classification\test'
```

## Load the Data

```python
batch_size = 64
img_size = 224

train_ds = image_dataset_from_directory(directory = train_data_dir, 
                                        labels = 'inferred', 
                                        label_mode = 'categorical',
                                        seed = 123,
                                        image_size = (img_size, img_size),
                                        batch_size = batch_size)


val_ds = image_dataset_from_directory(directory = val_data_dir, 
                                      labels = 'inferred', 
                                      label_mode = 'categorical',
                                      seed = 123,
                                      image_size = (img_size, img_size),
                                      batch_size = batch_size)


test_ds = image_dataset_from_directory(directory = test_data_dir, 
                                      labels = 'inferred', 
                                      label_mode = 'categorical',
                                      seed = 123,
                                      image_size = (img_size, img_size),
                                      batch_size = batch_size)

class_names = train_ds.class_names
print('\n', class_names)
```

## Create a function to change images names

```python
# The images have names like : 124242, 245235 etc...
# We dont like that, instead we want them to be tomato_0, tomato_1 etc...
# We actually dont mind for train and validation data, but we want to know what happens to test data

def change_name(directory):
    for vegetable in class_names:
        path = directory + '/' +  vegetable + '/' # Path for a file which contains images
        for i, image in enumerate(os.listdir(path)): # For each image in a certain path
            img = cv2.imread(path+str(image)) # Read the image
            if image.endswith(".jpeg") | image.endswith(".JPEG"):
                cv2.imwrite(path + vegetable + '_' + str(i) + str(image)[-5:], img) # Create a copy of that image with this name
            else:
                cv2.imwrite(path + vegetable + '_' + str(i) + str(image)[-4:], img)
            os.remove(path+image) # Remove that first image
```

## Lets build and train a basic CNN

```python
model = Sequential([
    layers.experimental.preprocessing.Rescaling(scale=1./255),
    Conv2D(filters=16, kernel_size=3, activation='relu'),
    MaxPool2D(),
    Conv2D(32, 3,  activation='relu'),
    MaxPool2D(),
    Conv2D(64, 3,  activation='relu'),
    MaxPool2D(),  
    Flatten(),
    Dense(64, activation='relu'),
    Dense(15, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

reduce_lr = ReduceLROnPlateau(
    monitor='accuracy', factor=0.9, patience=3, verbose=0, mode='auto', min_delta=0.01)

early_stop = EarlyStopping(
    monitor='accuracy', min_delta=0.01, patience=3, verbose=0, mode='auto')

history = model.fit(train_ds, batch_size=64, epochs=50, validation_data=val_ds, callbacks=[reduce_lr,early_stop])
```

## We make two functions to achive that

```python
# The first function gets an image and outputs an array
def image_to_array(file):
    img = tf.keras.preprocessing.image.load_img(file, target_size=(img_size, img_size, 3))
    img_array = tf.keras.preprocessing.image.img_to_array(img) 
    img_array = tf.expand_dims(img_array, 0)
    return img_array

# The second one gives the predictions
def get_predictions(model, directory):
    df = pd.DataFrame(columns=['File', 'prediction'])
    for veggie in class_names:
        for filename in os.listdir(directory + '/' + veggie): # For each image in a file in a directory

            try:
                image=image_to_array(directory + '/' + veggie + '/' + filename) # Get the array
                predictions = model.predict(image) # Predict
                score = tf.nn.softmax(predictions[0]) # Apply softmax
                # Add image name and which vegetable our model predicts
                df = pd.concat([df, pd.DataFrame({'File': [filename], 'prediction': [class_names[np.argmax(score)]]})], ignore_index=True)
                # Print image name and which vegetable our model predicts
                print(filename, 'prediction: ', class_names[np.argmax(score)])
            except:
                df = pd.concat([df, pd.DataFrame({'File': [filename], 'prediction': ["Not Possible"]})], ignore_index=True)
    
    # Now our image names are 'tomato_0', 'tomato_1' etc, and our predictions are 'tomato'
    df['Real_Category'] = df['File'].str.split('_').str[0] # Split image name and get the first part (vegetable name)
    df['Predicted_Category'] = df['prediction'] # Get the prediction
    
    # On 'Result' column we have 'Correct Prediction' where we have for ecample 'tomato'=='tomato'
    # And 'False Prediction' where we have for example 'tomato'=='cucumber'
    df['Result'] = np.where(df['Real_Category'].eq(df['Predicted_Category']), 'Correct Prediction', 'False Prediction')
    # Our function return how many correct and how many false predictions we had
    return df['Result'].value_counts()
```

# Train with Data Augmentation layers

#### We add data augmentation layers and also a dropout (we dont create new images with data augmentation at least for now)

```python
tf.keras.backend.clear_session()

model_aug = Sequential([
    tf.keras.layers.RandomFlip(),
    tf.keras.layers.RandomRotation(factor=0.05),
    tf.keras.layers.RandomContrast(0.05),
    tf.keras.layers.RandomZoom(0.2),
    layers.experimental.preprocessing.Rescaling(scale=1./255),
    Conv2D(filters=16, kernel_size=3, activation='relu'),
    MaxPool2D(),
    Conv2D(32, 3, activation='relu'),
    MaxPool2D(),
    Conv2D(64, 3, activation='relu'),
    MaxPool2D(),
    Dropout(0.2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(15, activation='softmax')
])

model_aug.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

reduce_lr = ReduceLROnPlateau(
    monitor='accuracy', factor=0.9, patience=3, verbose=0, mode='auto', min_delta=0.01)

early_stop = EarlyStopping(
    monitor='accuracy', min_delta=0.01, patience=3, verbose=0, mode='auto')

history_aug = model_aug.fit(train_ds, batch_size=64, epochs=50, validation_data=val_ds, callbacks=[reduce_lr,early_stop])
```

# Transfer Learning

### Lets try a pre-trained model like MobileNetV2

```python
tf.keras.backend.clear_session()

base_model = tf.keras.applications.MobileNetV2(weights='imagenet',
                                               include_top=False,
                                               input_shape=(img_size, img_size, 3))


base_model.trainable = False

model_tr = Sequential([
                    base_model,
                    layers.GlobalAveragePooling2D(),
                    layers.Dense(64, activation='relu'),
                    layers.Dense(15, activation='softmax')
                    ])

model_tr.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


reduce_lr = ReduceLROnPlateau(
    monitor='accuracy', factor=0.9, patience=3, verbose=0, mode='auto', min_delta=0.01)

early_stop = EarlyStopping(
    monitor='accuracy', min_delta=0.01, patience=3, verbose=0, mode='auto')

history_tr = model_tr.fit(train_ds, batch_size=64, epochs=50, validation_data=val_ds, callbacks=[reduce_lr,early_stop])
```

```python
model_tr.save("D:\\data_projects\\veggie_classification\\my_model_tr")

with open("D:\\data_projects\\veggie_classification\\history_tr_pickle", "wb") as file_pi:
    pickle.dump(history_tr.history, file_pi)
    
with open("D:\\data_projects\\veggie_classification\\history_tr_pickle", "rb") as file_pi:
    history_tr = pickle.load(file_pi)
    
model_tr = keras.models.load_model("D:\\data_projects\\veggie_classification\\my_model_tr")
```
